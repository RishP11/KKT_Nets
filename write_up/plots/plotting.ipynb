```python
import numpy as np 

import matplotlib.pyplot as plt 
import seaborn as sns
%config InlineBackend.figure_format='retina'
plt.rcParams.update({
    'font.size':10,
    'text.usetex':True
})

import tikzplotlib as tpl
```

Training Set Visualization 

```python
# Loading the data set
x_train = np.load('E:/Research/KKTNet/Source_Code_Simulations/QP/dataset/qp_train.npz')
print(x_train['x_opt_set'])
```

```python
x_opts = x_train['x_opt_set']
sample_points01 = x_opts[:10000]

fig, axes = plt.subplots()
sns.set_theme()
sns.scatterplot(sample_points01, x=sample_points01[:, 0], y=sample_points01[:, 1])
scale = 1
axes.set_xlim(-scale, scale)
axes.set_ylim(-scale, scale)
axes.set_xlabel('$x_{1}$')
axes.set_ylabel('$x_{2}$')
plt.savefig('data_set_viz.pdf')
```
